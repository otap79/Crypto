```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from config import schedule
from tqdm.notebook import tqdm
```

```python
DATA_PATH = "archive/"
folders = os.listdir(DATA_PATH)
folders.remove('.DS_Store')

data_dict = dict()
for folder in tqdm(folders):
    name = folder.split('.')[0]
    file2load = f"{DATA_PATH}{folder}{'/'}{folder}.gzip"
    df = pd.read_csv(file2load)
    df.index = pd.to_datetime(df['timestamp'])
    df = df.drop(columns = {'Unnamed: 0','timestamp','close_time'})
    df = df.sort_index()
    data_dict[name] = df
```

```python
Close = []
Volume = []

for ticker in tqdm(data_dict.keys()):
    Close.append(data_dict.get(ticker)['close'].rename(ticker))
    Volume.append(data_dict.get(ticker)['volume'].rename(ticker))

Close = pd.concat(Close,axis=1).ffill()
Volume = pd.concat(Volume,axis=1).fillna(0.0)
```
